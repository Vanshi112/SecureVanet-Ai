```python
import pandas as pd

rows = []

file_path = "/Users/vanshikasharma/Desktop/SecureVANET-AI/datasets/raw/normal_run_data.txt"   # Change if needed

with open(file_path, "r") as f:
    for line in f:

        line = line.strip()
        if not line:
            continue
        if not line.startswith("Timestamp:"):
            continue

        parts = line.split()

        try:
            timestamp = float(parts[1])
            can_id = int(parts[3], 16)
            rtr = int(parts[4])
            dlc = int(parts[6])
            data = parts[7:]
            data = [int(byte, 16) for byte in data]
            while len(data) < 8:
                data.append(0)
            data = data[:8]
            rows.append({
                "timestamp": timestamp,
                "can_id": can_id,
                "rtr": rtr,
                "dlc": dlc,
                "d0": data[0],
                "d1": data[1],
                "d2": data[2],
                "d3": data[3],
                "d4": data[4],
                "d5": data[5],
                "d6": data[6],
                "d7": data[7],
            })

        except Exception:
            continue

normal_dataset = pd.DataFrame(rows)

```

```python
from pathlib import Path
import pandas as pd

raw = Path("/Users/vanshikasharma/Desktop/SecureVANET-AI/datasets/raw/DoS_dataset.csv")
PROCESSED_DIR = Path("/Users/vanshikasharma/Desktop/SecureVANET-AI/datasets/processed")
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

rows = []

with open(raw, "r") as file:
    for line_num, line in enumerate(file, start=1):
        line = line.strip()
        if not line:
            continue

        parts = line.split(",")

        try:

            timestamp = float(parts[0])

            can_id = int(parts[1], 16)

            dlc = int(parts[2])
            payload = parts[3:-1]
            payload = [int(byte, 16) for byte in payload]
            while len(payload) < 8:
                payload.append(0)

            payload = payload[:8]
            raw_label = parts[-1].strip()

            if raw_label == "R":
                label = 0
                attack_type = "normal"

            elif raw_label == "T":
                label = 1
                attack_type = "dos"

            else:
                continue

            rows.append({
                "timestamp": timestamp,
                "can_id": can_id,
                "rtr": 0,
                "dlc": dlc,
                "d0": payload[0],
                "d1": payload[1],
                "d2": payload[2],
                "d3": payload[3],
                "d4": payload[4],
                "d5": payload[5],
                "d6": payload[6],
                "d7": payload[7],
                "label": label,
                "attack_type": attack_type
            })

        except Exception as e:

            print(f"Skipping line {line_num}: {e}")

            continue
dos_dataset = pd.DataFrame(rows)
output_file = PROCESSED_DIR / "dos_processed.csv"
dos_dataset.to_csv(output_file, index=False)
```
